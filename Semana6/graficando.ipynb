```python
## dibujando funciones
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-2,2,500) # 500 puntos equiespaciados entre el mínimo y el máximo y lo guarda en un array de numpy

y = x**2 

plt.plot(x, y)

# poniéndolo bonito

plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
```

```python
plt.show()
```

```python
# Cuando la tengamos a nuestro gusto la podemos guardar como fichero

plt.savefig("my_square_function.png", transparent=True)
```
